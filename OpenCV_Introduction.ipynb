**Reading an Image :**

- PIL - Python Image Library

````python
import cv2
````

````python
img = cv2.imread('image.jpg')
````

````python
type(img)
````

Data type of img variable is numpy array (as image is made up of array of pixels)

````python
img.shape
````

````python
# plotting the image

import matplotlib.pyplot as plt
plt.imshow(img)
````

````python
from  google.colab. patches import cv2_imshow

img2 = cv2.resize(img, (400, 225))
cv2_imshow(img2)
````

````python
cv2_imshow(sliced_img)
````
